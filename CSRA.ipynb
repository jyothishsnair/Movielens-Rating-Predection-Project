```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
 Data=pd.read_csv("Customer Service Requests Analysis.csv")
```

```python
Data.info()
```

```python
Data.columns
```

```python
DD=['Unique Key', 'Created Date', 'Closed Date', 'Agency', 'Agency Name',
       'Complaint Type', 'Descriptor', 'Location Type',
       'City','Facility Type', 'Status', 'Due Date',
       'Resolution Description', 'Resolution Action Updated Date',
       'Community Board', 'Borough', 'X Coordinate (State Plane)',
       'Y Coordinate (State Plane)','Latitude', 'Longitude', 'Location']
FinalData=Data[DD]
FinalData.info()
```

```python
FinalData['Closed Date'].fillna(FinalData['Closed Date'].mode()[0],inplace=True)
FinalData['Descriptor'].fillna(FinalData['Descriptor'].mode()[0],inplace=True)
FinalData['Location Type'].fillna(FinalData['Location Type'].mode()[0],inplace=True)
FinalData['City'].fillna(FinalData['City'].mode()[0],inplace=True)
FinalData['Facility Type'].fillna(FinalData['Facility Type'].mode()[0],inplace=True)
FinalData['Due Date'].fillna(FinalData['Due Date'].mode()[0],inplace=True)
FinalData['Resolution Action Updated Date'].fillna(FinalData['Resolution Action Updated Date'].mode()[0],inplace=True)
FinalData['X Coordinate (State Plane)'].fillna(FinalData['X Coordinate (State Plane)'].mean(),inplace=True)
FinalData['Y Coordinate (State Plane)'].fillna(FinalData['Y Coordinate (State Plane)'].mean(),inplace=True)
FinalData['Latitude'].fillna(FinalData['Latitude'].mean(),inplace=True)
FinalData['Longitude'].fillna(FinalData['Longitude'].mean(),inplace=True)
FinalData['Location'].fillna(FinalData['Location'].mode()[0],inplace=True)
```

```python
FinalData.info()
```

```python
FinalData['Created Date']=pd.to_datetime(FinalData['Created Date'])
FinalData['Closed Date']=pd.to_datetime(FinalData['Closed Date'])
```

```python
FinalData['Request_Closing_Time']=FinalData['Closed Date']-FinalData['Created Date']
```
